```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="darkgrid")
```

```python
#LOAD DATASET
df=pd.read_csv(r"C:\Users\anshi\OneDrive\Desktop\ZOMATO EDA PROJECT\zomato.csv")
print(df.head())
```

```python
#EXPLORING DATASET INFORMATION
print(df.info())
print(df.shape)
print(df.columns)
```

```python
print(df.isnull().sum())
```

```python
#CLEANING NULL VALUES
df.drop_duplicates()
df['phone']=df['phone'].fillna("not availaible")
df['location']=df['location'].fillna("unknown")
df['rest_type']=df['rest_type'].fillna("unknown")
for i in ['dish_liked','cuisines']:
    if not df[i].mode().empty:
        df[i]=df[i].fillna(df[i].mode()[0])
    else:
        df[i]=df[i].fillna("unknown")
df['approx_cost(for two people)']=df['approx_cost(for two people)'].astype(str)
df['approx_cost(for two people)']=df['approx_cost(for two people)'].apply(lambda x:x.replace(",",""))
df['approx_cost(for two people)']=df['approx_cost(for two people)'].astype(float)
df['approx_cost(for two people)']=df['approx_cost(for two people)'].fillna(df['approx_cost(for two people)'].median())
df['rate']=df['rate'].astype(str)
df['rate']=df['rate'].apply(lambda x:x.replace("/5","") if '/5' in x else x)
df['rate']=df['rate'].replace(["NEW","-"],np.nan)
df['rate']=df['rate'].astype(float)
df['rate']=df['rate'].fillna(df['rate'].mode()[0])

print(df.isnull().sum())
```

```python
#FORMATING THE TEXT COLUMNS
df['cuisines']=df['cuisines'].str.strip()
df['location']=df['location'].str.strip()
df['rest_type']=df['rest_type'].str.strip()
```

```python
#NORMALIZE CATEGORIES
df['online_order']=df['online_order'].map({'Yes':1,'No':0})
df['book_table']=df['book_table'].map({'Yes':1,'No':0})

```

```python
#Q1--TOP 10 LOCATIONS WITH MOST RESTAURANTS
top_countries=df['location'].value_counts().head(10)
print(top_countries)

sns.barplot(x=top_countries.values,y=top_countries.index)
plt.xlabel("NUMBER OF RESTAURANTS")
plt.ylabel("LOCATION")
plt.title("TOP 10 LOCATIONS WITH MOST RESTAURANTS")
plt.savefig("images/top_10_locations.png", dpi=300, bbox_inches='tight')

plt.show()
```

```python
#Q2--ONLINE ORDER VS OFFLINE ORDER

online_order_count=df['online_order'].value_counts()
print(online_order_count)

pl=sns.countplot(x="online_order",data=df,hue="online_order",palette=["green","darkblue"],legend=False)
for container in pl.containers:
    pl.bar_label(container, label_type="edge")
plt.title("ONLINE_ORDER(1) VS OFFLINE_ORDER(0)")
plt.savefig("images/order_type.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
#Q3--TABLE BOOKING VS NO BOOKING
table_booked_count=df['book_table'].value_counts()
print(table_booked_count)

pl=sns.countplot(x="book_table",data=df,hue="book_table",palette=["yellow","orange"],legend=False)
for container in pl.containers:
    pl.bar_label(container, label_type="edge")
plt.title("TABLE_BOOKED(1) VS TABLE_NOT_BOOKED(0)")
plt.savefig("images/table_booking.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
#Q4--MOST POPULAR CUISINES
df['cuisines_list']=df['cuisines'].str.split(',')
cuisines_explode=df.explode('cuisines_list')
cuisines_explode['cuisines_list']=cuisines_explode['cuisines_list'].str.strip()
top_cuisines=cuisines_explode['cuisines_list'].value_counts().head(10)
print(top_cuisines)

sns.barplot(x=top_cuisines.values,y=top_cuisines.index,hue=top_cuisines.values,palette="Set2")
plt.title("MOST POPULAR CUISINES")
plt.savefig("images/popular_cuisines.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
#Q5--RATING DISTRIBUTION(HISTOGRAM OF RESTAURENTS RATING)
rating_count=df['rate'].value_counts()
print(rating_count)
sns.displot(df["rate"])
plt.savefig("images/rating_distribution.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
#Q6--VOTES VS RATING (DO HIGHER RATING GETS MORE VOTE ?)

sns.scatterplot(x='votes', y='rate', data=df,color="green")
plt.savefig("images/votes_vs_rating.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
#Q7-COST FOR TWO ANALYSIS (BOXPLOT OF APPROX_COST VS LISTED_IN(TYPE))
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].astype(str)
df['approx_cost(for two people)']=df['approx_cost(for two people)'].str.replace(',',' ').astype(float)
sns.boxplot(x="approx_cost(for two people)",y="listed_in(type)",data=df,hue="listed_in(type)",palette="Set2")
plt.title("APPROX_COST VS LISTED_IN(TYPE)")
plt.savefig("images/cost_for_two.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
#Q8--WHICH LOCATION HAS THE BEST AVERAGE RATING?
avg_rating=df.groupby('location')['rate'].mean()
best_avg_rating = avg_rating.sort_values(ascending=False).head(5)
print(best_avg_rating)
sns.barplot(x=best_avg_rating.values, y=best_avg_rating.index,hue=best_avg_rating.index, palette="viridis")
plt.xlabel("RATING AVERAGE")
plt.ylabel("LOCATION")
plt.title("LOCATION HAVING MORE AVERAGE RATING")
plt.savefig("images/rating_according_to_location.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
#Q9--WHICH TYPE OF RESTAURENT DOMINATE?
df['rest_type_list']=df['rest_type'].str.split(',')
rest_type_explode=df.explode('rest_type_list')
rest_type_explode['rest_type_list']=rest_type_explode['rest_type_list'].str.strip()
rest_type_dominate=rest_type_explode['rest_type_list'].value_counts()
print(rest_type_dominate)

sns.barplot(
    x=rest_type_dominate.head(10).values,
    y=rest_type_dominate.head(10).index,
    hue=rest_type_dominate.head(10).values,
    palette="viridis",
    legend=False
)
plt.xlabel("Count")
plt.ylabel("Restaurant Type")
plt.title("Top 10 Dominating Restaurant Types")
plt.savefig("images/dominating_type.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
#Q10--RESTAURENT DISTRIBUTION BY LISTED_IN(CITY)
restaurant_count=df['listed_in(city)'].value_counts()
print(restaurant_count)
sns.barplot(
    x=restaurant_count.values,
    y=restaurant_count.index,
    hue=restaurant_count.values,
    palette="viridis"
)
plt.xlabel("Number of Restaurants")
plt.ylabel("City")
plt.title("Restaurant Distribution by City")
plt.savefig("images/restaurant_distribution_by_city.png", dpi=300, bbox_inches='tight')
plt.show()
```
